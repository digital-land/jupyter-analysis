```python
import os
import urllib
import fiona
from fiona import Feature, Geometry
from shapely.geometry import mapping, shape
import shapely.wkt
```

```python
# download flattened csv as an example
data_dir = '../data/geopackage_maker'
os.makedirs(data_dir,exist_ok=True)
data_path = os.path.join(data_dir,'brownfield-land.csv')
urllib.request.urlretrieve(
    'https://files.planning.data.gov.uk/dataset/brownfield-land.csv',
    data_path
)
```

```python
with fiona.open(
    data_path
) as src:

    # The file we'll write to must be initialized with a coordinate
    # system, a format driver name, and a record schema. We can get
    # initial values from the open source's profile property and then
    # modify them as we need.
    profile = src.profile
    print(src.profile)
    profile["schema"]["geometry"] = "Point"
    profile["driver"] = "GPKG"

    del profile["schema"]["properties"]["geometry"]
    del profile["schema"]["properties"]["point"]
    print(profile)


    # Open an output file, using the same format driver and coordinate
    # reference system as the source. The profile mapping fills in the
    # keyword parameters of fiona.open.
    geopackage_path = os.path.join(data_dir,'brownfield-land.gpkg')
    with fiona.open(geopackage_path, "w", **profile) as dst:
        # Process only the feature records intersecting a box.
        for feat in src:
            # Get the feature's centroid.
            # print(feat.properties['geometry'])
            
            point_wkt = feat.properties['point']
            print(point_wkt)
            if point_wkt:
                point = Geometry.from_dict(shapely.wkt.loads(feat.properties['point']))
                del feat.properties['point']
                del feat.properties['geometry']
                dst.write(
                    Feature(geometry=point, properties=feat.properties)
                )
            else :
                point = None
            # # new_geom = Geometry.from_dict(centroid_shp)
            # print(point)
            # Write the feature out.
            
            

    # The destination's contents are flushed to disk and the file is
    # closed when its with block ends. This effectively
    # executes ``dst.flush(); dst.close()``.
```
