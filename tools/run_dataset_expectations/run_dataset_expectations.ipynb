## Running a yaml data validation configuration
This is a short notebook put together to show how to use the expectations tool inside of digital land python. We have mainly based these off the sqllite databases for each dataset, these are all avaulable via the datasettes which sit behind the platform.

The first cell allows you to set where results will be stored, the location of the thq data which is being validated. (in this. case in. a sqlite db) 

```python
from download_data import download_dataset,download_dataset_expectations_yaml

from digital_land.expectations.main import run_expectation_suite
from digital_land.expectations.suite import DatasetExpectationSuite

import os
import pandas as pd
# from digital_land.expectations.main import run_dq_suite
```

```python
#  define dataset to run expectations for as also the directory to store data in
dataset = 'conservation-area'
collection = 'conservation-area-collection'

# use these to set other values
dataset_file_name = f'{dataset}.sqlite3'
dataset_yaml_file_name = f'{dataset}.yaml'
data_dir = os.path.join('../data/run_dataset_expectations',dataset)
dataset_path = os.path.join(data_dir,dataset_file_name)
dataset_yaml_path = os.path.join(data_dir,dataset_yaml_file_name)
results_file_path = os.path.join(data_dir,f'{dataset}-results.csv')
```

```python
# download dataset
download_dataset(dataset,collection,data_dir)
download_dataset_expectations_yaml(dataset,collection,data_dir)
```

```python
# use to run and access in memory
expectation_suite = DatasetExpectationSuite(
        results_file_path, dataset_path, dataset_yaml_path
    )
expectation_suite.run_suite()
responses = [response.to_dict() for response in expectation_suite.responses]
```

```python
responses_df = pd.DataFrame.from_records(responses)
responses_df
```

```python
# use this to save to folder
run_expectation_suite(results_file_path, dataset_path, dataset_yaml_path)
```
