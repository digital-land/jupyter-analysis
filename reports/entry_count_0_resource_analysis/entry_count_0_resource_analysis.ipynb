```python
import pandas as pd
import urllib
```

```python
datasette_url = "https://datasette.planning.data.gov.uk/"
result_df = pd.DataFrame()

def get_endpoint_resource_entry_info():
    params = urllib.parse.urlencode({
        "sql": f"""
       select
  organisation,
  name,
  organisation_name,
  dataset,
  collection,
  pipeline,
  endpoint,
  endpoint_url,
  status,
  days_since_200,
  resource
from
  reporting_latest_endpoints
where (collection == 'article-4-direction' OR collection == 'conservation-area' OR collection == 'listed-building' OR collection =='tree-preservation-order')
        """,
        "_size": "max"
    })
    
    url = f"{datasette_url}digital-land.csv?{params}"
    df = pd.read_csv(url)
    return df
```

```python
datasets= ['article-4-direction', 'article-4-direction-area', 'conservation-area', 'listed-building-outline', 'tree', 'tree-preservation-order', 'tree-preservation-zone' ]
```

```python
def get_entry_count_per_dataset():
    datasette_url = "https://datasette.planning.data.gov.uk/"
    all_resources = []  # List to store all resources
    
    for dataset in datasets:
      dataset_database_url = f"{datasette_url}{dataset}"
      print(dataset_database_url)
    
      params = urllib.parse.urlencode({
            "sql": f"""
                select
                  entry_count,
                  resource
                from
                  dataset_resource
                where entry_count == 0
                        """,
            "_size": "max"
        })
    
       
      url = f"{dataset_database_url}.csv?{params}"
      df = pd.read_csv(url) 
      all_resources.extend(df[['entry_count', 'resource']].values.tolist())
        
    resources_df = pd.DataFrame(all_resources, columns=['entry_count', 'resource'])
    return resources_df
    

```

```python
resources_df = get_entry_count_per_dataset()
resources_df
```

```python
pd.merge(entry, resources_df, on='resource', how='inner').drop_duplicates()
```
